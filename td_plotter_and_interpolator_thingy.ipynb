```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('ss_td.csv')
```

```python
raw
```

```python
raw['int vel (ft/s)'][0] = 0
raw['ave vel (ft/s)'][0] = 0
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(raw['tvd (ft)'], raw['2 way time (s)'])

plt.title('SS 45 T-D Relation')
ax.set_xlabel('depth (ft)')
ax.set_ylabel('2 way time (s)')

ax.set_yticks(range(5))
ax.set_xticks(range(0, 23000, 1000))
plt.minorticks_on()
plt.grid(b=True, which='major', color='black', linestyle='-')
plt.grid(b=True, which='minor', color='grey', linestyle='-')

plt.show()
```

```python
fig = plt.figure(figsize=(24,24), dpi=500)
ax = fig.add_subplot(111)
plt.plot(raw['tvd (ft)'], raw['2 way time (s)'])

plt.title('SS 45 T-D Relation')
ax.set_xlabel('depth (ft)')
ax.set_ylabel('2 way time (s)')

ax.set_yticks([0.5,1,1.5,2,2.5,3,3.5,4,4.5,5])
ax.set_xticks(range(0, 23000, 1000))
plt.minorticks_on()
plt.grid(b=True, which='major', color='black', linestyle='-')
plt.grid(b=True, which='minor', color='grey', linestyle='--')

plt.savefig('ss_td.png')
```
